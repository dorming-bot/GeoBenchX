# SKILL Benchmark Notebook

通过本 Notebook 可以在不改动主仓库基准脚本的情况下，直接调用 `skill_test` 目录下的独立生成器，批量求解任务集并输出 JSON + HTML。

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / "geobenchx").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

TASK_FILE = PROJECT_ROOT / "skill_test/dataset/skill_tasks_and_reference_solutions.json"
RESULTS_DIR = PROJECT_ROOT / "skill_test/skill_benchmark/results"
OUTPUT_FILENAME = "skill_generated_solutions.json"
MODEL = "gpt-4o"
TEMPERATURE = 0.0
MAX_STEPS = 25
CAPTURE_HISTORY = True  # 生成 HTML 对话
FORCE_RERUN = False     # True 时会重新求解已有结果的任务
```

```python
from geobenchx.dataclasses import TaskSet
from skill_test.skill_benchmark.generate_solution import generate_solutions

tasks = TaskSet.read_from_file(TASK_FILE.name, folder=str(TASK_FILE.parent))
generate_solutions(
    tasks=tasks,
    model=MODEL,
    temperature=TEMPERATURE,
    output_filename=OUTPUT_FILENAME,
    max_steps=MAX_STEPS,
    skip_solved=not FORCE_RERUN,
    capture_history=CAPTURE_HISTORY,
    results_folder=str(RESULTS_DIR),
)
print("Done. Results saved under:", RESULTS_DIR)
```
